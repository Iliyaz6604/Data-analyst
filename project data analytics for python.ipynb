```python
### Region analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gb = df.groupby(["Region"]).agg({"Profit": "sum", "Sales" : "sum","Discount": "sum"}).reset_index()
print(gb)
```

```python
# here we need profit margin so we caliculated the (profit / sales)*100
#later we got the values as default it is ascending so we need to change it to false to get best values first
#so we sorted vales and set it for ascending 
```

```python
gb["Profit_margin"] = (gb["Profit"]/gb["Sales"])*100
print(gb.sort_values("Profit", ascending =False))
```

```python
#visualization of the data
```

```python
plt.figure(figsize=(8,7))
sns.barplot(data = gb, x = "Region", y = "Profit", palette = 'viridis')
plt.title("Region wise profit")
plt.show()
```

```python
##From above data we can conclude that in WEST region the profit is high
```

```python
"""customer segment analysis"""
```

```python
gbc = df.groupby(["Segment"]).agg({"Profit": "sum", "Sales" : "sum", "Order ID": "nunique"}).reset_index()
print(gbc)
```

```python
gbc_margin = ((gbc["Sales"]/gbc["Order ID"])*100)
print(gbc_margin)
```

```python
plt.figure(figsize=(5, 3))
sns.barplot(x='Segment', y='Profit', data=gbc)
plt.title('Profit by Customer Segment')
plt.show()
```

```python
'''Consumer Segment: Highest total profit and sales.
Corporate Segment: Moderate profit with higher AOV'''
```

```python
gbsub = df.groupby(["Sub-Category","Category"]).agg({"Profit": "sum", "Sales" : "sum", "Order ID": "nunique","Discount": "mean"}).reset_index()
print(gbsub.sort_values("Profit", ascending= False))
```

```python
gbsub_avg = (gbsub["Sales"]/gbsub['Order ID'])*100
print(gbsub_avg.sort_values(ascending= False))
```

```python
plt.figure(figsize=(18,5))
sns.barplot(data =gbsub,y='Profit',x = 'Sub-Category' )
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.barplot(data =gbsub,y='Profit',x = 'Category' )
plt.show()
```

```python
product_analysis = df.groupby('Product Name').agg({'Profit': 'sum', 'Sales': 'sum', 'Discount': 'mean'}).reset_index()
top_products = product_analysis.sort_values('Profit', ascending=False).head(10)
worst_products = product_analysis.sort_values('Profit').head(25)
print("Top Products", top_products)
print("Worst Products", worst_products)
```
